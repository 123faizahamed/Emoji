```python
from skimage.io import imread , imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
sad = os.listdir("Sad")
```

```python
angry = os.listdir("Angry")
```

```python
sun = os.listdir("Sunglass")
```

```python
limit = 10
sad_images = [None] * limit
j = 0 

for i in sad:
    if j < limit:
        if i.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            sad_images[j] = imread("Sad//" + i)
            j += 1
    else:
        break
```

```python
limit = 10
angry_images = [None] * limit
j = 0 

for i in angry:
    if j < limit:
        if i.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            angry_images[j] = imread("Angry//" + i)
            j += 1
    else:
        break
```

```python
limit = 10
sun_images = [None] * limit
j = 0 

for i in sun:
    if j < limit:
        if i.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            sun_images[j] = imread("Sunglass//" + i)
            j += 1
    else:
        break
```

```python
sad_gray = [None]*limit # sad - rajini
j = 0 
for i in sad:
    if(j<limit):
        sad_gray[j] = rgb2gray(sad_images[j][:,:,:3])
        j+=1
    else:
        break
```

```python
sun_gray = [None]*limit # sun vijay 
j = 0 
for i in sun:
    if(j<limit):
        sun_gray[j] = rgb2gray(sun_images[j][:,:,:3])
        j+=1
    else:
        break
```

```python
angry_gray = [None]*limit # angry Dhanush
j = 0 
for i in angry:
    if(j<limit):
        angry_gray[j] = rgb2gray(angry_images[j][:,:,:3])
        j+=1
    else:
        break
```
